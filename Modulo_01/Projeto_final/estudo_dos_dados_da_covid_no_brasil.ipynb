####Criando as colunas de registros diários

Para os dados acumulados, criarei duas novas colunas, uma que registra o número de óbitos naquele dia e o outro como número de novos casos registrados na mesma data. Pois para analisar a média móvel, é necessário saber o registro diário.

```
# O que foi feito aqui foi:
#1. definir a ordem do dataframe para que seja ordenado por estado e order_for_place, que registra o valor ordinal daquele registro
#2. Agrupar por estado, de forma que a função aconteça somente naquele estado.
#3. usar somente as series de casos confirmados
#4. usar a função diff, que calcula a diferença entre o valor de cima com o valor abaixo e atribui ao valor abaixo.
#5. Passar o valor para valor absoluto, para não obter valores negativos.
#6. Atribuição essa função acima à uma variavel chamada grupo2 (eu achei que já tinha usado uma chamado grupo, mas na vdd não tinha xD)

grupo2 = abs(covid_estado.sort_values(["state","order_for_place"], ascending =False).groupby("state").confirmed.diff())
```

```
# Verificação de como ficou a variavél grupo2

grupo2
```

Essa series está correta porém está na ordem "errada" para o que eu queria, pois como não existe um valor acima do ultimo registro não tem com o que fazer a função diff. O que eu queria é que o valor 222 estivesse no index 5780 (que no caso é a primeira linha), e que todos os valores subissem um valor também, e ai todos os primeiros valores do new_confirmed fossem o primeiro valor do confirmed.

Vou tentar fazer isso tudo nas próximas linhas.

```
# Ordenar o DataFrame igual da series criada para ser o new_confirmed_daily que foi criada
covid_estado = covid_estado.sort_values(["state","order_for_place"], ascending =False)


# O que fiz aqui:
#1. Criei a variável primeira_entrada para receber ultimo valor da series Confirmed, que é o primeiro valor de registro do estado do acre
#2. Redefini a series grupo2, de forma que o primeiro valor (NaN) fosse pulado. E como tenho mais facilidade usando listas
#transformei a variavel grupo2 em lista e peguei todos menos o primeiro dado dela.
#3. Usei a função append para adicionar ao final da lista o valor do confirmed do primeiro registro do Acre.
#se isso não fosse feita a lista teria menos valores do que o index do dataframe a qual ele seria introduzido,
#assim não seria possível atribuir a lista à nova series que seria a nova coluna. 
primeira_entrada = covid_estado.confirmed.to_list()[-1]
grupo2 = list(grupo2[1:])
grupo2.append(primeira_entrada)
```

```
# Visualização dos primeiros 5 dados da lista

grupo2[0:6]
```

```
# Atribuição dos dados à nova coluna e reordenação do index

covid_estado["new_confirmed_daily"] = grupo2
covid_estado.index = [x for x in range(covid_estado.shape[0])]
```

####Criação das colunas de média móvel

Agora quero criar uma coluna de média móvel nos Dataframes de dados cumulativos. A média móvel é uma forma que foi encontrada de "normalização" dos dados, de registros diários, pois a variação diária de valores não ajudava no estudo da evolução da doença, então a utilização da média móvel ajuda a compreender a média de variação dos dados nos ultimos 7 dias, como é comunmente usado.

```
# Para fazer essas novas colunas vou utilizar algumas coisas usadas na criação de casos diários e outras novas

# A função "principal" aqui é o .rolling ele cria uma janela de seleção de dados para fazer o calculo que for necessário.
#Nesse sentido novamente ordenei os dados, porém dessa vez de ordem crescente, pois o rolling calcula de "cima para baixo" em uma series.
#Novamente os dados foram agrupados para não se misturarem no calculo. Selecionei os casos diários e usei a função transform
#assim ela pega esses valores da series e transforma de acordo com a função que eu usar, que no caso
#foi fazer média então dessa janela de 7 dias definida pelo uso do rolling.

media_movel_casos = covid_estado.sort_values(by=["state","order_for_place"],
                                             ascending=True).groupby("state").new_confirmed_daily.transform(lambda x: x.rolling(7).mean())
media_movel_casos
```
